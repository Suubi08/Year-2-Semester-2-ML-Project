```python
import pandas as pd
import glob
```

```python
#look for all csv files that start with "air" in the folder
files = glob.glob('air-*.csv')

#combine them into one single dataframe
df = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)

#sort by time so the sequence is correct
df['datetime'] = pd.to_datetime(df['datetime'])
df = df.sort_values(by='datetime')

#saving the master data
df.to_csv('Kampala_air_quality_master.csv', index=False)
print("All batches combined successfully")
```
